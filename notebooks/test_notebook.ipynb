```python
college = pd.read_csv(DATA_PATH + 'college.csv')
college.rename(columns={'Unnamed: 0': 'College'}, inplace=True)
college.set_index('College', inplace=True)
college
```

```python
perc_cols = [col for col in college.columns if 'perc' in col]
college[perc_cols] = college[perc_cols] / 100
college['Elite'] = np.where(college['Top10perc'] > 0.5, 'Yes', 'No')

college['Elite'].value_counts(dropna=False)
```
